```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("result_data/steps_oct_2021.csv")
data = data.drop(columns={'Unnamed: 0.1', 'Unnamed: 0'})
print(data.columns)
print(data.shape)
data.head()
```

```python
## Normalize the columns
def normalize_99(column):

    p99 = np.percentile(data[column].dropna(), 99)
    data[column] = np.where(data[column] > p99, p99, data[column])

    min_val = data[column].min()
    max_val = data[column].max()

    data[column] = (data[column] - min_val) / (max_val - min_val)


columns_to_normzalize = data[['t2m', 'mcc', 'lcc', 'hcc', 'u10']]

for column in columns_to_normzalize:
    normalize_99(column)

data.head()
```
